```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#loading all cvs
temperatures = pd.read_csv('WebScrapping/src/main/temperatures.csv')
humidity = pd.read_csv('WebScrapping/src/main/humidity.csv')
windspeed = pd.read_csv('WebScrapping/src/main/windspeed.csv')
pressure = pd.read_csv('WebScrapping/src/main/pressure.csv')
```

```python
#creating one df
df= pd.concat([temperatures, humidity, windspeed, pressure], axis=1)
print(df)
```

```python
df.plot(kind='scatter',x='humidity',y='temperatures')
df.plot(kind='scatter',x='windspeed',y='temperatures')
df.plot(kind='scatter',x='pressure',y='temperatures')
```

```python
#creating X dataset and y dataset
y = df.iloc[:,0].values
X = df.iloc[:,1:4].values
```

```python
# Split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
# Perform linear regression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
# root mean squared error
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')


```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='True')
plt.plot(y_pred, label='Predicted')
plt.legend(loc='upper left')
plt.title('Comparison of predicted values and y test values')
plt.show()
```
