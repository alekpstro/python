```python
#import data about workers of menagment
#LECTURE 4, EXAMPLE 3 
# step 1
page=requests.get("https://wzr.ug.edu.pl/wydzial/index.php?str=121")
```

```python
# step 2 extract the data 
from bs4 import BeautifulSoup
tags=BeautifulSoup(page.text,'html.parser')
employees=tags.find_all('div',{'class':'pracownik_b'})
#step 3- opening the file to write
file = open("employees_284509.csv","w") #w means write r read
for employee in employees:
    print(employee.text.strip())
#step 3 write one line to the file
    file.write(employee.text.strip()+"\n")
file.close()
```

```python
#step 2- extract the data
tags=BeautifulSoup(page.text,'html.parser')
products=tags.find_all('span',{'class':'notranslate plp-price-module__product-name'})
#<span class="notranslate plp-price-module__product-name">VINTERFINT</span>
for product in products:
    print(product.text.strip())
#<span class="plp-price__sr-text">59,49</span>
prices=tags.find_all('span',{'class':'plp-price__sr-text'})
for price in prices:
    print(price.text.strip())
```

```python
# step 2 - extract the data
tags=BeautifulSoup(page.text, 'html.parser')
products=tags.find_all('span',{'class':'pip-header-section__title--small notranslate'})
for product in products:
    print(product.text.strip())

prices=tags.find_all('span',{'class':'pip-temp-price__nowrap'})
for price in prices:
    print(price.text.strip())
```
