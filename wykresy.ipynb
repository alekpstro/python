```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
df = pd.read_csv("realestate_prices.csv", decimal=',', delimiter=";")
df.plot(kind='scatter', x="sqm", y='price')
```

```python
df.head()
```

```python
df.shape
```

```python
 #1 st case - detect prices based on squared meters 
X=np.c_[df["sqm"]]
y=np.c_[df["price"]]
model=linear_model.LinearRegression()
#train the model
model.fit(X,y)
```

```python
# we will predict the price based on squared meters
print(model.predict([[31]])) # flat 31 sqm
print(model.predict([[51]])) # flat 51 sqm
```

```python
# 2 nd case - we will add number of rooms
X = np.c_[df['sqm'],df['rooms']]
y = np.c_[df['price']]
model=linear_model.LinearRegression()
#train the model
model.fit(X,y)
```

```python
# predicting the price 
print(model.predict([[31,4]])) # 31sqm and 4 rooms
print(model.predict([[66,3]]))
print(model.predict([[101,7]]))
```
